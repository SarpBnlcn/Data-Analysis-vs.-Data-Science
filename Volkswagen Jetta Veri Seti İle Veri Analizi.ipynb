```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns
import seaborn as sbn
```

```python
veri = pd.read_csv("C:/Users/benli/OneDrive/Masaüstü/Jetta.csv")
veri.head()
```

Sütun İsimlerini Değiştirme ve NaN Değerin Olduğu Satırı Silme İşlemi

```python
veri = pd.DataFrame(veri)
veri.rename(columns={'Column 1': 'Fiyat', 'Column 2': 'Birim'}, inplace=True)
veri.drop(index=3, inplace=True)
veri
```

Veri Setinin Düzenlenmesi

```python
veri['Fiyat'] = veri['Fiyat'].str.replace(' TL', '', regex=False)
veri['Birim'].fillna("TL", inplace=True)
veri
```

Fiyatlar Sütununu Float Veri Tipine Dönüştürme

```python
veri['Fiyat'] = veri['Fiyat'].str.replace(' TL', '').str.replace('.', '').astype(float)
```

Veri Görselleştirme - Box Plot

```python
plt.figure(figsize=(10, 6))
plt.boxplot(veri['Fiyat'])
plt.title(' Box Plot')
plt.ylabel('Fiyat (TL)')
plt.show()
```
